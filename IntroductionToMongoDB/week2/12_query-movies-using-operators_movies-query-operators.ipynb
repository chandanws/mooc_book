```python
import pymongo
import pprint
from pymongo import MongoClient
```

```python
mc = MongoClient('mongodb://analytics:analytics-password@mflix-shard-00-00-mz7dv.mongodb.net:27017, mflix-shard-00-01-mz7dv.mongodb.net:27017, mflix-shard-00-02-mz7dv.mongodb.net:27017/?ssl=true&replicaSet=mflix-shard-0&authSource=admin')
mflix = mc['mflix']
```

```python
# find all movies released from 1983 onwards
filters = {'year': { '$gte': 2010 }}

for movie in mflix.movies.find(filters):
    pprint.pprint(movie['title'])
```

```python
# find all movies between 1989 and 1999
filters = { 'year': {'$gte': 1989, '$lt': 2000} }

for movie in mflix.movies.find(filters):
    pprint.pprint(movie['title'])
```

```python
# find all movies released in 1995, 2005, 2015
filters = { 'year': { '$in': [ 1995, 2005, 2015 ] } }
for movie in mflix.movies.find(filters):
    pprint.pprint(movie['title'])
```

```python
# find all movies except the ones which are of adult genre
filters = { 'year': { '$in': [ 1995, 2005, 2015 ] }, 
                     'genre': { '$not' : {'$eq': 'Adult'} } }

for movie in mflix.movies.find(filters):
    pprint.pprint(movie['title'])
```
