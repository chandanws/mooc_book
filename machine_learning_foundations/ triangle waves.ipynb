# triangle waves


Consider the "triangle waves'' hypothesis set on $\mathcal{R}$, which is given by
$\mathcal{H}=\{h_\alpha|h_\alpha(x)=\text{sign}(|(\alpha x) \text{mod} 4−2|−1),\alpha \in \mathcal{R}\}$
Here $(z \text{mod} 4)$ is a number $z−4k$ for some integer $k$ such that $z−4k\in [0,4)$. For instance, (11.26 mod 4) is 3.26, and (−11.26 mod 4) is 0.74. What is the VC-dimension of such an $\mathcal{H}$?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def triangle_waves(x, alpha=1):
    return np.sign(np.abs((alpha*x)%4-2)-1)
```

```python
x = -np.arange(-10,10,0.1)
y = triangle_waves(x, alpha=1)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
